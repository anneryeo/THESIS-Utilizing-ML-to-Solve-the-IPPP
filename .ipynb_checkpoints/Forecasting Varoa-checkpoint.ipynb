```python
import zipfile
import pandas as pd
```

```python
path_to_zip_file = r"C:\Users\rjbel\Python\Data\Thesis\Predicting Varroa Longitudinal Data, Micro Climate, and Proximity Closeness Useful for Predicting Varroa Infestations.zip"
directory_to_extract_to = r"C:\Users\rjbel\Python\Data\Thesis"
path_of_csv_files = "Predicting Varroa Longitudinal Data, Micro Climate, and Proximity Closeness Useful for Predicting Varroa Infestations"

csv_hive = rf"{directory_to_extract_to}\{path_of_csv_files}\hive.csv"
csv_station = rf"{directory_to_extract_to}\{path_of_csv_files}\station.csv"
csv_user = rf"{directory_to_extract_to}\{path_of_csv_files}\user.csv"
csv_varroa_sampling = rf"{directory_to_extract_to}\{path_of_csv_files}\varroa_sampling.csv"
csv_weather = rf"{directory_to_extract_to}\{path_of_csv_files}\weather.csv"
csv_yard = rf"{directory_to_extract_to}\{path_of_csv_files}\yard.csv"
```

```python
# Extract zip file
with zipfile.ZipFile(path_to_zip_file, 'r') as zip_ref:
  zip_ref.extractall(directory_to_extract_to)


# Read csv files
df_hive = pd.read_csv(csv_hive)
df_station = pd.read_csv(csv_station)
df_user = pd.read_csv(csv_user)
df_varroa_sample = pd.read_csv(csv_varroa_sampling)
df_weather = pd.read_csv(csv_weather)
df_yard = pd.read_csv(csv_yard)
```
